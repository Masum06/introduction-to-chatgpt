## Guess what word comes next!!

````python
import sys
````



```
story = "Dear ..."
```



````python
#@title
story = "Dear Professor I know it sounds weird but my dog ate my homework and that's why I am late can I get an extension"
````

````python
tokens = story.split()
num_trials = []
````

````python
print("Total Number of words: ", len(tokens))
````

````python
print("Type exit() to quit.\n\nFirst word: Dear\n\n")

for token in tokens[1:]:
  guess = ""
  i = 0
  while True:
    i += 1
    guess = input().strip().lower()
    if guess == "exit()":
      print("Byee!! 🤗")
      sys.exit(0)
    if guess == token.lower():
      num_trials.append(i)
      print(">>", "Correct!! 😄 Number of trial:", i, "\n")
      break
    else:
      print(">>", "Wrong!! 🥺 Number of trial:", i,  "\n")
````

````python
for (i, j) in zip(num_trials, tokens[:len(num_trials)]):
  print(i, j)
````
